```python
#import dependencies

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from api_keys import weather_api_key
import os
from scipy import stats
from citipy import citipy
import time
from scipy.stats import linregress
```

```python
#set variables url and query_url to include the weather open weather url and use that url to make a query url 
#so that we can pull data from open weather map
url = "https://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"http://api.openweathermap.org/data/2.5/weather?appid={weather_api_key}&units={units}&q="
```

```python
#jsonify the query_url
weather_response = requests.get(query_url)
weather_json = weather_response.json()

#latitude and longitude ranges

latitude_range = (-90,90)
longitude_range = (-180,180)

#generate variable to hold a list for coordinates

coordinates = []

#np.random.uniform draws samples from a uniform distribution and uses low and high as array_like floats
#size in this argument is used to define an integer or tuple of integers
random_latitude = np.random.uniform(low=-90, high=90, size=1500)
random_longitude = np.random.uniform(low=-90, high=90, size=1500)

#fill blank coordinates list with randomly generated latitudes and longitudes
coordinates = zip(random_latitude, random_longitude)

#generate variable called cities and countries to hold a list

cities = []
countries = []

#use for loops to loop through and identify citipy using the coordinates list
#nearest_city and city_name can be found in the source code of citipy https://pydoc.net/citipy/0.0.5/citipy.citipy/
for coordinates in coordinates:
    city = citipy.nearest_city(coordinates[0], coordinates[1]).city_name
    country = citipy.nearest_city(coordinates[0], coordinates[1]).country_code
    
#find unique values in the city variable and append those values to the cities list    
    if city not in cities:
        cities.append(city)
        countries.append(country)
        
len(cities)
```

```python
#Perform Api Calls

#Create Lists

latitude = []
longitude = []
weather_date = []
temp = []
humidity = []
cloudiness = []
wind_speed = []
city_variable = []
countries_2 = []

#variables for counter values to go through

i_counter = 0
counter = 1

for city in cities:
    try:
        
        response = requests.get(query_url + city.replace(" ","&")).json()
        
        countries_2.append(response['sys']['country'])
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
        weather_date.append(response['dt'])
        temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
    
        if i_counter > 48:
            i_counter = 1
            counter += 1
            city_variable.append(city)
        else:
            i_counter += 1
            city_variable.append(city)
        print(f"Processing Record {i_counter} of Set {counter} | {city}")
        
    except Exception:
        print("City not found. Skipping...")
        
        
print("------------------------------\nData Retrieval Complete\n------------------------------")
```

```python
weather_dictionary = {
    "City":city_variable,
    "Cloudiness":cloudiness,
    "Country":countries_2,
    "Date":weather_date,
    "Humidity":humidity,
    "Lat":latitude,
    "Lng":longitude,
    "Max Temp": temp,
    "Wind Speed": wind_speed
    }
```

```python
weather_df = pd.DataFrame(weather_dictionary)

weather_df.head()
```

```python
#define northern and southern hemisphere values
northern = weather_df.loc[weather_df["Lat"] >= 0.0]
northern.reset_index(inplace=True)

southern = weather_df.loc[weather_df["Lat"] < 0.0]
southern.reset_index(inplace=True)
```

```python
#Linear Regression for Northern Hemisphere Max Temp vs Latitude
north_temp = northern["Max Temp"]
north_lat = northern["Lat"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(north_temp, north_lat)
regress_values = north_temp * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(north_temp, north_lat)
plt.ylim(-5,100)
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.plot(north_temp,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

print(f"The r-value is {rvalue}")
```

```python
#Southern Hemisphere Max Temp vs Latitude
south_temp = southern["Max Temp"]
south_lat = southern["Lat"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(south_temp, south_lat)
regress_values = south_temp * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(south_temp, south_lat)
plt.ylim(-75,50)
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.plot(south_temp,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

print(f"The r-value is {rvalue}")
```
